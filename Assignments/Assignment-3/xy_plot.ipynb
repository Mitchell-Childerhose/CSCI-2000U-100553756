```python
#Mitchell Childerhose (100553756)

from matplotlib import *
from pylab import *
import numpy as np
import matplotlib.patches as mpatches
%matplotlib inline

figure(3, figsize=(8,8))

with open('xy_points.dat', 'r') as f:
    first_line = f.readline()
    second_line = f.readline()

x = first_line.split()
y = second_line.split()

x = [float(i) for i in x]
y = [float(j) for j in y]

xlist = [i for i in x if any(i <= -2)]
ylist = [j for j in y if any(-5 <= j <= 0)]      

scatter(x,y, color = 'red')
scatter(xlist,xlist, color = 'blue')
scatter(ylist,ylist, color = 'blue')

xticks(range(-10,11), color = 'black')
yticks(range(-10,11), color = 'black')

title('Highlighted points (100553756)')
ylabel('Y', color = 'black', rotation = 0)
xlabel('X', color = 'black')

xlim(-10, 10)
ylim(-10, 10)

show()
```
