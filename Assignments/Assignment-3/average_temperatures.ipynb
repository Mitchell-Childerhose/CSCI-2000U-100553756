```python
#Mitchell Childerhose (100553756)

from matplotlib import * 
from pylab import *
import numpy as np
import matplotlib.patches as mpatches
%matplotlib inline

#creating the figure
figure(3, figsize=(8,6))

#reading the montreal_temperature_2012.dat file
montreal = np.loadtxt('montreal_temperature_2012.dat',
                 unpack=True,
                 delimiter = '\n')

#reading the vancouver_temperature_2012.dat file
toronto = np.loadtxt('vancouver_temperature_2012.dat',
                 unpack=True,
                 delimiter = '\n')

#reading the toronto_temperature_2012.dat file
vancouver = np.loadtxt('toronto_temperature_2012.dat',
                 unpack=True,
                 delimiter = '\n')

#setting 3 letter abbreviations on the x axis
labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
x = range(0,12)

plot1 = plot(x,montreal, linewidth = 3, color = 'red')
plot2 = plot(x,toronto, linewidth = 3, color = 'blue')
plot3 = plot(x,vancouver, linewidth = 3, color = 'black')

xticks(x, labels, rotation=45, color = 'black')
yticks(color = 'black')

#adding the title 
title('2012 average monthly temperatures (100553756)')
ylabel('Temperature (C)', color = 'black')
xlabel('Month', color = 'black')

#adding the legend
red_patch = mpatches.Patch(color='red', label='Montreal')
blue_patch = mpatches.Patch(color='blue', label='Toronto')
black_patch = mpatches.Patch(color='black', label='Vancouver')
legend(handles=[red_patch,blue_patch,black_patch])

show()
```
